```python
# Make directory if it doesn't already exist
folder_name = 'image-predictions'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(url)
response
```

```python
# open and write image predictions
with open(os.path.join(folder_name, url.split('/')[-1]), mode='wb') as file:
    file.write(response.content)
```

```python
# verify that the file is there
os.listdir(folder_name)
```

```python
# keys
CONSUMER_KEY = ''
CONSUMER_SECRET = ''
ACCESS_KEY = ''
ACCESS_SECRET = ''

auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_KEY, ACCESS_SECRET)

api = tweepy.API(auth)

```

```python
# store twitter archive as df
df = pd.read_csv('twitter-archive-enhanced.csv')

# get tweets from df
tweet_id = list(df['tweet_id'])

# list and dictionary to store tweet info
df_list = []
tweet_errors = {}
count = 1
start = time.time()

with open('tweet_json.txt', 'w') as f:
    for tweet in tweet_id:
        try:
            tweet = api.get_status(tweet, tweet_mode='extended', wait_on_rate_limite=True, wait_on_rate_limit_notify=True)
            # save to tweet_json.txt file
            f.write(str(tweet.id) + ', ' + str(tweet.retweet_count) + ', ' + str(tweet.favorite_count) + '\n')
            # append to list of dictionaries
            df_list.append({'tweet_id': tweet.id,
                            'retweet_count': tweet.retweet_count,
                            'favorite_count': tweet.favorite_count})

        except Exception as e:
            tweet_errors[count] = tweet
            count += 1

end = time.time()
print(end - start)
```

<b>Note for the reviewer:</b> I am getting a lot of tweets in the Exception dictionary. 1400 tweets in total. Is this normal? 

```python
# create DataFrame from list of dictionaries
df_tweets = pd.DataFrame(df_list, columns = {'tweet_id', 'retweet_count', 'favorite_count'})
df_tweets.head()
```

# Cleaning the data

```python
df_clean = df.copy()
predictions_clean = predictions.copy()
df_tweets_clean = df_tweets.copy()
```

## Issue 1: Merging all the DataFrames (tidiness issue #1)

<b><i>Define</i></b>

Merge all DataFrames in order to conform with tidiness principles.

<b><i>Code</i></b>

```python
# merge df_tweets and df
df_clean = pd.merge(df_clean, df_tweets, left_on = 'tweet_id', right_on = 'tweet_id', how = 'inner')

# merge predictions and df
df_clean = pd.merge(df_clean, predictions, left_on = 'tweet_id', right_on = 'tweet_id', how = 'inner')
```

<b><i>Test</i></b>

```python
df_clean.head()
```

```python
df_clean.shape
```

##  Issue 2: Remove tweets that don't have images

<b><i>Define</i></b>

Remove tweets rows that don't have images. This means removing rows where jpg_url is null.

<b><i>Code</i></b>

```python
# find the number of tweets that don't have an image
df_clean.jpg_url.isnull().sum()
```

```python
# find the number of tweets that have an image
len(df_clean[df_clean.jpg_url.notnull()])
```

```python
# remove tweets that don't have an image
df_clean = df_clean[df_clean.jpg_url.notnull()]
```

<b><i>Test</i></b>

```python
df_clean.shape
```

```python
# Check to see if expanded_urls has null values, which would mean we will have tweets with no pictures
df_clean['expanded_urls'].isnull().sum()
```

## Issue 3: Remove all retweets

<b><i>Define</i></b>

Remove all the retweets. We know there is a retweet when `retweeted_status_id` for example is not null. Let's remove all these rows. 

<b><i>Code</i></b>

```python
# find the number of tweets that are retweets
df_clean.retweeted_status_id.notnull().sum()
```

```python
# find the number of tweets that are not retweets
df_clean.retweeted_status_id.isnull().sum()
```

```python
# remove tweets that are retweets
df_clean = df_clean[df_clean.retweeted_status_id.isnull()]
```

<b><i>Test</i></b>

```python
df_clean.shape
```

## Issue 4: Drop all irrelevent columns (tidiness issue #2)

<b><i>Define</i></b>

Drop all the columns related to retweets since they are not useful - they only contain the null value. 

<b><i>Code</i></b>

```python
# drop the 3 columns related to retweets 
df_clean = df_clean.drop(['retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp', 'in_reply_to_status_id', 'in_reply_to_user_id', 'p2', 'p2_conf', 'p2_dog', 'p3', 'p3_conf', 'p3_dog'], axis=1)
```

<b><i>Test</i></b>

```python
list(df_clean.columns)
```

## Issue 5: Fix denominators 

<b><i>Define</i></b>

Remove all rating_denominator values that are not 10. Since this happens when there is another fraction in the text, we'll simply remove these rows for simplicity's sake. Not many rows are affected so this won't be a problem. 

<b><i>Code</i></b>

```python
# Take only denominators that are equal to 10
df_clean = df_clean[df_clean['rating_denominator'] == 10]
```

<b><i>Test</i></b>

```python
# check to see if there are denominators not equal to 10
df_clean[df_clean['rating_denominator'] != 10]
```

## Issue 6: Fix numerators

<b><i>Define</i></b>

Remove all rating_numerators that are over 20. They are either errors, or ratings with denominators that are not equal to 10. They will skew any visual or statistic so we're better off removing them. We'll also remove when the numerator is 0. 

<b><i>Code</i></b>

```python
# take all numerators that are less than or equal to 20, and not equal to 0
df_clean = df_clean[df_clean['rating_numerator'] <= 20]
df_clean = df_clean[df_clean['rating_numerator'] != 0]
```

<b><i>Test</i></b>

```python
df_clean['rating_numerator'].value_counts()
```

## Issue 7: Fix dog names 

<b><i>Define</i></b>

Change dog names that have lowercase names to 'None'

<b><i>Code</i></b>

```python
# convert lower case names to 'None'
lower_names = df['name'].loc[(df.name.str.islower())]
for name in lower_names:
    df = df.replace(name, 'None')
```

<b><i>Test</i></b>

```python
df.loc[(df.name.str.islower())]
```

```python
df.name.value_counts()
```

## Issue 8: Fix dtypes

<b><i>Define</i></b>

`df`: Change `timestamp`, `tweet_id`, `source` columns' dtypes

<b><i>Code</i></b>

```python
df_clean.timestamp = pd.to_datetime(df_clean.timestamp)
df_clean.tweet_id = df_clean.tweet_id.astype(str)
df_clean.source = df_clean.source.astype(str)
```

## Issue 9: Fix source column

<b><i>Define</i></b>

Replace the full URL in the `source` column to only the name of the source.

<b><i>Code</i></b>

```python
# catch the url pattern with regex and isolate the text of the link
df_clean.source = df_clean.source.str.extract('([^<a href="http://twitter.com/download/iphone" rel="nofollow">].+[^</a>])')
```

# Visualisation and Insights

Let's see if there is a correlation between the rating_numerator and the retweet count or the favorite count. 

Before we do this, we need to check how many of each numerators there are to know if we include them or not in our analysis. 

```python
df_clean.rating_numerator.value_counts()
```

We'll only include numerators 10-14 because the other numerators are only represented once or twice. 

```python
# Get the average favorite count of tweets with ratings higher than 10
fav = df_clean[df_clean['rating_numerator'] >= 10].groupby('rating_numerator')['favorite_count'].mean()
fav
```

```python
retweet = df_clean[df_clean['rating_numerator'] >= 10].groupby('rating_numerator')['retweet_count'].mean()
retweet
```

```python
# Get ratings higher than 10
df_clean2 = df_clean[df_clean['rating_numerator'] >= 10]

# plot rating_numerator and favorite_count
f, ax = plt.subplots(figsize=(23, 9))
sns.barplot(x=df_clean2.rating_numerator, y=df_clean.favorite_count)
ax.set_title('Average favorited tweet per rating', fontsize=40)
ax.set_xlabel('Ratings', fontsize=30)
ax.set_ylabel('Average favorited tweet', fontsize=30)
plt.show()
```

### Insight #1: 
The higher the rating given to the dog picture, the more people tended to favorite the image. 

```python
# plot rating_numerator and retweet_count
f, ax = plt.subplots(figsize=(23, 9))
sns.barplot(x=df_clean2.rating_numerator, y=df_clean.retweet_count)
ax.set_title('Average number of retweets per rating', fontsize=50)
ax.set_xlabel('Ratings', fontsize=30)
ax.set_ylabel('Average number of retweets', fontsize=30)
plt.show()
```

### Insight #2: 
The higher the rating given to the dog picture, the more people tended to retweet the image. This was very evident for ratings 13 and 14.  

Let's have a look at the results of the algorithm. Which dog races and objects was it better at predicting? 

```python
# get the prediction confidence levels for all dogs and objects
prob = df_clean.groupby('p1')['p1_conf'].mean().sort_values(ascending=False).reset_index()
high_prob = prob[prob['p1_conf'] >= 0.8]
high_prob
```

```python
# plot the top 30 dogs/objects
f, ax = plt.subplots(figsize=(23, 30))
sns.barplot(x = 'p1_conf', y = 'p1', data=high_prob)
ax.set_title('Prediction confidence per predictor', fontsize=50)
ax.set_xlabel('Prediction confidence', fontsize=30)
ax.set_ylabel('Predictor', fontsize=30)
plt.show()
```

### Insight #3:
The algorithm seems to be much better at predicting objects since most of the top 10 predictors are objects.
